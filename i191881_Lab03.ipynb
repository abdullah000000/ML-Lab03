```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x = np.array([7,6])
# create orthogonal vector to x
y = np.array([-x[1], x[0]])
y
    



```

```python
a = x.dot(y)
a
```

```python
# Take a projection of x onto y and plot it
x_proj = (a / y.dot(y)) * y
x_proj
# plot the vectors
plt.plot([0, x[0]], [0, x[1]], 'r', label='x')
plt.plot([0, y[0]], [0, y[1]], 'b', label='y')
plt.plot([0, x_proj[0]], [0, x_proj[1]], 'g', label='x_proj',marker='o')
plt.legend()
plt.axis('square')
plt.show()



```

```python
#now find projection of y onto x
y_proj = (a / x.dot(x)) * x
y_proj
#plot the vectors
plt.plot([0, x[0]], [0, x[1]], 'r', label='x')
plt.plot([0, y[0]], [0, y[1]], 'b', label='y')
plt.plot([0, y_proj[0]], [0, y_proj[1]], 'g', label='y_proj',marker='o')
plt.legend()
plt.axis('square')
plt.show()
```

```python
# Now find the initial vectors from the projections
x_init = x_proj + y_proj
x_init
```

```python
y_init = y_proj + x_proj
y_init
```

```python
# Take two vectors and project one on the other, and find the initial vectors from the projections.
# Plot the vectors and the projections
plt.plot([0, x[0]], [0, x[1]], 'r', label='x')
plt.plot([0, y[0]], [0, y[1]], 'b', label='y')
plt.plot([0, x_proj[0]], [0, x_proj[1]], 'g', label='x_proj',marker='o')
plt.plot([0, y_proj[0]], [0, y_proj[1]], 'y', label='y_proj',marker='v')
plt.plot([0, x_init[0]], [0, x_init[1]], 'k', label='x_init',marker='^')
plt.plot([0, y_init[0]], [0, y_init[1]], 'm', label='y_init',marker='.')
plt.legend()
plt.axis('square')
plt.show()
```
